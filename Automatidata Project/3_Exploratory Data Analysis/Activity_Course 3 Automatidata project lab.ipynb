This plot indicates that the drop off points are relatively uniformly distributed over the total area covered.

To confirm this conclusion, consider the following experiment:

Create a sample of coordinates from a normal distribution—in this case 1,500 pairs of points from a normal distribution with a mean of 10 and a standard deviation of 5
Calculate the distance between each pair of coordinates
Group the coordinates by endpoint and calculate the mean distance between that endpoint and all other points it was paired with
Plot the mean distance for each unique endpoint


```python
# 1. Generate random points on a 2D plane from a normal distribution
test = np.random.normal(10, 5, (3000, 2))
midway = int(len(test)/2)  # Calculate midpoint of the array of coordinates
start = test[:midway]      # Isolate first half of array ("pick-up locations")
end = test[midway:]        # Isolate second half of array ("drop-off locations")

# 2. Calculate Euclidean distances between points in first half and second half of array
distances = (start - end)**2           
distances = distances.sum(axis=-1)
distances = np.sqrt(distances)

# 3. Group the coordinates by "drop-off location", compute mean distance
test_df = pd.DataFrame({'start': [tuple(x) for x in start.tolist()],
                   'end': [tuple(x) for x in end.tolist()],
                   'distance': distances})
data = test_df[['end', 'distance']].groupby('end').mean()
data = data.sort_values(by='distance')

# 4. Plot the mean distance between each endpoint ("drop-off location") and all points it connected to
plt.figure(figsize=(14,6))
ax = sns.barplot(x=data.index,
                 y=data['distance'],
                 order=data.index)
ax.set_xticklabels([])
ax.set_xticks([])
ax.set_xlabel('Endpoint')
ax.set_ylabel('Mean distance to all other points')
ax.set_title('Mean distance between points taken randomly from normal distribution');
```

This graph is very similar to one we obtained for previously. This suggests that our Location points are close to normally distributed.

# **Histogram of rides by drop-off location**

First, check to whether the drop-off locations IDs are consecutively numbered. For instance, does it go 1, 2, 3, 4..., or are some numbers missing (e.g., 1, 3, 4...). If numbers aren't all consecutive, the histogram will look like some locations have very few or no rides when in reality there's no bar because there's no location. 

```python
# Check if all drop-off locations are consecutively numbered
df['DOLocationID'].max() - len(set(df['DOLocationID']))  #set() returns all unique values in the list
```

To eliminate the spaces in the historgram that these missing numbers would create, sort the unique drop-off location values, then convert them to strings. This will make the histplot function display all bars directly next to each other. 

```python
plt.figure(figsize=(14,6))
# DOLocationID column is numeric, so sort in ascending order
sorted_locationID = df['DOLocationID'].sort_values()
# Convert to string
sorted_locationID = sorted_locationID.astype('str')
# Plot
sns.histplot(sorted_locationID, bins = range(0, df['DOLocationID'].max()+1))
plt.xticks([])
plt.xlabel('Drop off Locations')
plt.title('Histogram of rides by Drop off location')
```

We can see that there are certain drop off locations with very high volume of rides. Getting more information regarding these locations could be useful.

Let us plot a correlation heatmap of all features along with trip duration to select the most important features for model training

```python
# Calculate the correlation matrix
#Let's take only data that has numerical values and drop nominal and ordinal data
columns_to_drop = [0,1,2,3,6,7,8,9,10,18,19]
df_tmp = df.drop(df.columns[columns_to_drop], axis = 1)

#Let's also drop other categorical features encoded as integer as they will give misleading information in corr matrix
correlation_matrix = df_tmp.corr()

# Plotting the correlation heatmap using Seaborn
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

**What can we infer from this correlation heatmap?**

We see that trip_distance is highly related to total_amount, which is our independent variable. The other features that have high correlation with total_fare are just other types of charges which are calculated based of fare amount. Hence, those features can be dropped out of our modelling.
The other features that impact total amount but not very significantly are passenger_count and trip_distance. If these lead to high computation cost and do not add much accuracy, we can drop these features.
